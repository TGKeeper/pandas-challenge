# PyCity Schools Analysis

The analysis considers some very general factors in relation to student performance in math and reading.  The data set consists of 15 schools along with their budgets, student numbers, grade levels, and student scores.  As part of the assignment, the data was reorganized several times to highlight factors such as the per student budget, different grade levels, whether it was a charter or district shool, and the size of the student population.   

At first glance, overall pass rates seem correlated to school size.  The 7th largest schools, with student numbers closer to the 3,000- to 5,000-student range, not only sit at the bottom range, they trail as much as 38 percentage points behind the better performing schools. The data seems to divide in roughly half, with overall pass rates in either the 90s or the 50s.  

Since chool size is a widely publicized performance factor and an easy concept for any lay person to grasp, latching onto that part of the data would make sense if you were, for instance, lobbying for new school construction or a similar initiative.  The problem is the diffference between Wilson & Ford High Schools.  Their student populations differ by 456 students, so less than 20%, but they have almost the widest gap in student achievement within the data set.  Only slightly more than half of Ford HS students are passing their courses while Wilson HS students pass at a rate of better than 9 out of 10, even though they rank lowest in per student spending. Granted, maybe that 20% difference represents the tipping point of inadequate resources to student numbers, and given that it is only one school, maybe the difference between Ford & Wilson high schools should just be discounted as an anomaly.  Presumably other factors are effecting the pass rates more, such as teacher to student ratios, student economic backgrounds, how outliers such as AP students with weighted averages and students with learning disabilities are handled within the reporting data, etc.  

I also expected a strong connection between student performance and the school's per student budget.  However, within this small data set at least, the schools in the lower half of student spending outperformed their better funded competitors significantly, with only the exception of Thomas HS. Again, within just this very small data set, bigger schools appear to spend more and still achieve far less.
